**Collaborative Filtering**

Now, time to start our work on recommendation systems.

The first technique we're going to take a look at is called Collaborative Filtering, which is also known as User-User Filtering. As hinted by its alternate name, this technique uses other users to recommend items to the input user. It attempts to find users that have similar preferences and opinions as the input and then recommends items that they have liked to the input. There are several methods of finding similar users (Even some making use of Machine Learning), and the one we will be using here is going to be based on the Pearson Correlation Function.



The process for creating a User Based recommendation system is as follows:

* Select a user with the movies the user has watched
* Based on his rating to movies, find the top X neighbours
* Get the watched movie record of the user for each neighbour.
* Calculate a similarity score using some formula
* Recommend the items with the highest score

```python
userInput = [
            {'Name':'Zocalo Mexican Grill & Tequileria', 'rating':7},
            {'Name':'Barrio', 'rating':8},
            {'Name':'HoB Restaurant & Bar', 'rating':6},
            {'Name':"Rally's", 'rating':4},
            {'Name':'Sushi 86', 'rating':7}
         ] 
inputRestaurants = pd.DataFrame(userInput)
inputRestaurants
```

```python
#Filtering out the restaurants by name
inputId = cle_rest[cle_rest['Name'].isin(inputRestaurants['Name'].tolist())]
#Then merging it so we can get the RestaurantId. It's implicitly merging it by title.
inputRestaurants = pd.merge(inputId, inputRestaurants)

#Final input dataframe
#If a movie you added in above isn't here, then it might not be in the original 
#dataframe or it might spelled differently, please check capitalisation.
inputRestaurants
```

```python
#Filtering out the movies from the input
userRestaurants = cle_rest[cle_rest['Restaurant ID'].isin(inputRestaurants['Restaurant ID'].tolist())]
userRestaurants
```

```python
#Resetting the index to avoid future issues
userRestaurants = userRestaurants.reset_index(drop=True)
#Dropping unnecessary issues due to save memory and to avoid issues
userRatings = userRestaurants.drop('Restaurant ID', 1).drop('Name', 1).drop('Latitude', 1).drop('Longitude', 1).drop('id', 1)
userRatings
```

```python
inputRestaurants['Rating']
```

```python
#Dot product to get weights
userProfile = userRestaurants.transpose().dot(inputRestaurants['rating'])
#The user profile
userProfile
```

```python
#Now let's get the genres of every movie in our original dataframe
ratingsTable = cle_rest.set_index(cle_rest['Restaurant ID'])
#And drop the unnecessary information
ratingsTable = ratingsTable.drop('Restaurant ID', 1).drop('Name', 1).drop('Latitude', 1).drop('Longitude', 1).drop('id', 1)
ratingsTable.head()
```

```python
ratingsTable.shape
```

```python
#Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((ratingsTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```
